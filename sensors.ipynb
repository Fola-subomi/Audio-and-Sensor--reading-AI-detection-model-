```python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
Sensor_audio = ''

for filename in os.listdir(Sensor_audio):
    if filename.endswith('.wav'):
        audio_path = os.path.join(Sensor_audio, filename)
        
        y, sr = librosa.load(audio_path, sr=16000)

        print(f"Sampling rate: {sr}")
        print(f"Audio length (seconds): {len(y)/sr:.2f}")
```

```python
for file in os.listdir(Sensor_audio):
    if file.endswith(".wav"):   # or .mp3
        audio_path = os.path.join(Sensor_audio, file)

        # Load audio
        y, sr = librosa.load(audio_path, sr=16000)

        time = np.linspace(0, len(y) / sr, num=len(y))

        # Plot waveform
        plt.figure(figsize=(12, 4))
        plt.plot(time, y)
        plt.xlabel("Time (seconds)")
        plt.ylabel("Amplitude")
        plt.title(f"Waveform: {file}")
        plt.show()

        # Compute Mel spectrogram
        mel_spec = librosa.feature.melspectrogram(
            y=y,
            sr=sr,
            n_fft=1024,
            hop_length=512,
            n_mels=64
        )

        # Convert to dB
        mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

        # Plot Mel spectrogram
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(
            mel_spec_db,
            sr=sr,
            hop_length=512,
            x_axis="time",
            y_axis="mel"
        )
        plt.colorbar(format="%+2.0f dB")
        plt.title(f"Mel Spectrogram: {file}")
        plt.tight_layout()
        plt.show()
```
